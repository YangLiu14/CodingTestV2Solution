## 53. 在排序数组中查找数字

题目一：数字在**排序数组**中出现的次数。

统计一个数字在排序数组中出现的次数。例如，输入排序数组[1,2,3,3,3,3,4,5]和数字3，由于3在这个数组中出现了4次，因此输出4.

### 分析
“排序”数组，很容易让人想到二分查找算法。先用二分法找到中间的3（一次性就找到第一个出现的3和最后一个出现的3，这种情况很简单，不多讨论）。先往左找用二分法找到第一个3，再往右找用二分法找到最后一个3.

```python
# ===========================
# 找到k第一次出现时的位置
# ===========================

# 递归方法
def get_first_k(arr: list, k: int, start: int, end: int):
    if start > end:
        return -1

    mid_index = (end + start) // 2

    if arr[mid_index] == k:
        if mid_index == 0:
            return 0
        if mid_index > 0 and arr[mid_index - 1] != k:
            return mid_index
        else:
            # go left
            end = mid_index - 1

    elif arr[mid_index] > k:
        # go left
        end = mid_index - 1
    else:
        # go right
        start = mid_index + 1

    return get_first_k(arr, k, start, end)


# 循环方法
def get_first_k_loop(arr: list, k: int):
    if arr is None or arr == []:
        return -1

    start = 0
    end = len(arr) - 1
    mid_index = (end + start) // 2

    while True:
        if start > end:
            return -1
        if mid_index == 0 and arr[0] == k:
            return 0

        if arr[mid_index] == k:
            if arr[mid_index - 1] == k:
                # go left
                end = mid_index - 1
            else:
                return mid_index
        elif arr[mid_index] > k:
            # go left
            end = mid_index - 1
        else:
            # go right
            start = mid_index + 1

        mid_index = (end + start) // 2

        
# Unit Test
arr1 = [1, 2, 3, 3, 3, 3, 4, 5]
print("should print 2")
print(get_first_k(arr1, 3, 0, len(arr1)-1))
print(get_first_k_loop(arr1, 3))

arr2 = [4, 5]
print("should print -1")
print(get_first_k(arr2, 3, 0, len(arr2)-1))
print(get_first_k_loop(arr2, 3))

arr3 = [3, 3, 4, 5]
print("should print 0")
print(get_first_k(arr3, 3, 0, len(arr3)-1))
print(get_first_k_loop(arr3, 3))

arr4 = [1, 2, 3]
print("should print 2")
print(get_first_k(arr4, 3, 0, len(arr4)-1))
print(get_first_k_loop(arr4, 3))
```

```python
# ===========================
# 找到k最后一次出现时的位置
# ===========================

# 递归方法
def get_last_k(arr: list, k: int, start: int, end: int):
    if start > end:
        return -1

    mid_index = (end + start) // 2
    if arr[mid_index] == k:
        if mid_index == len(arr) - 1:
            return len(arr) - 1
        if arr[mid_index + 1] > k:
            return mid_index
        else:
            # go right
            start = mid_index + 1
    elif arr[mid_index] < k:
        # go right
        start = mid_index + 1
    else:
        # go left
        end = mid_index - 1

    return get_last_k(arr, k, start, end)


# 循环方法
def get_last_k_loop(arr: list, k: int):
    if arr is None or arr == []:
        return -1
    start = 0
    end = len(arr) - 1
    mid_index = (end + start) // 2

    while True:
        if start > end:
            return -1
        if mid_index == len(arr) - 1 and arr[mid_index] == k:
            return len(arr) - 1

        if arr[mid_index] == k:
            if arr[mid_index + 1] > k:
                return mid_index
            else:
                # go right
                start = mid_index + 1
        elif arr[mid_index] < k:
            # go right
            start = mid_index + 1
        else:
            # go left
            end = mid_index - 1

        mid_index = (end + start) // 2
        
# Unit Test
arr1 = [1, 2, 3, 3, 3, 3, 4, 5]
print("should print 5")
print(get_last_k(arr1, 3, 0, len(arr1)-1))
print(get_last_k_loop(arr1, 3))

arr2 = [4, 5]
print("should print -1")
print(get_last_k(arr2, 3, 0, len(arr2)-1))
print(get_last_k_loop(arr2, 3))

arr3 = [3, 3, 4, 5]
print("should print 1")
print(get_last_k(arr3, 3, 0, len(arr3)-1))
print(get_last_k_loop(arr3, 3))

arr4 = [1, 2, 3]
print("should print 2")
print(get_last_k(arr4, 3, 0, len(arr4)-1))
print(get_last_k_loop(arr4, 3))
```

```python
# ===========================
# 算出k出现的个数
# ===========================

def get_num_of_k(arr: list, k: int):
    number = 0
    if arr is not None and len(arr) > 0:
        first = get_first_k_loop(arr, k)
        last = get_last_k_loop(arr, k)
        
        if first > -1 and last > -1:
            number = last - first + 1
    
    return number

# Test
arr1 = [1, 2, 3, 3, 3, 3, 4, 5]
print("should print 4")
print(get_num_of_k(arr1, 3))

arr2 = [4, 5]
print("should print 0")
print(get_num_of_k(arr2, 3))

arr3 = [3, 3, 4, 5]
print("should print 2")
print(get_num_of_k(arr3, 3))

arr4 = [1, 2, 3]
print("should print 1")
print(get_num_of_k(arr4, 3))
```
